# Examples of using the DasDennis class to generate weight vectors

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
import das_dennis
```

## Example: 
- partitions: 12
- dimension: 3

```python
das_dennis = das_dennis.DasDennis(12, 3)
```

```python
print(das_dennis.get_number_of_points())
```

```python
points = das_dennis.get_weight_vectors() 
```

```python
print(points)
```

## Plot with Matplotlib

```python

fig = plt.figure()
 
# syntax for 3-D projection
ax = plt.axes(projection ='3d')
 
# defining axes
x = [p[0] for p in points]
y = [p[1] for p in points]
z = [p[2] for p in points]
ax.scatter(x, y, z)
 
# syntax for plotting
ax.set_title('3D plot')
plt.show()
```
