```python
import numpy as np
```

```python
def factorial(n):
    if (n == 0 or n == 1):
        return 1
    else: 
        return n * factorial(n-1)
```

```python
print(factorial(0))
print(factorial(1))
print(factorial(3))
```

```python
def binomial_coefficient(n, k):
    return factorial(n) / (factorial(k) * factorial(n - k))
```

## CASE ONE: H = 5, m = 3

```python
H = 5
m = 3
```

```python
print(binomial_coefficient(H + m - 1, m - 1))
```

```python
# FIRST LEVEL

first_level = np.linspace(0,1,H+1)
print("First level: " + str(first_level))
step_size = 1 / H
print("Step size: " + str(step_size))

for i in range(0, H+1):
    print (i * step_size )

print([i*step_size for i in range(0,H+1)])
```

```python
# SECOND LEVEL
index = 0
second_level = []
for i in first_level:
    #print("Value: " + str(i) + ". Index: " + str(index))
    values = [first_level[j] for j in range(len(first_level) - index)] 
    second_level.append((i, values))
    #print(values)

    index = index + 1
    
for i in second_level:
    print(i)
```

```python
# THIRD LEVEL
index = 0
third_level = []
for i in second_level:
    #print(i)
    for j in i[1]:
        third_level.append([i[0],j, 1.0 - j - i[0]])

print("N: " + str(len(third_level)))        
for i in third_level:
    print(i)
```
